```python
#Abdel

#system
import os
import sys

#DataScience
import pandas as pd    #pandas

#analisis numerico
import numpy as np

#librerias graficas
import matplotlib.pyplot as plt
import seaborn as sns

#M l  ----"Machine Learning"
import tensorflow as tf
import keras
import sklearn
from sklearn.model_selection import train_test_split

#import el dataset de datos
#url1 = "/content/drive/MyDrive/Datasets/dataset.csv" #leer de internet
url1 = "https://drive.google.com/uc?export=download&id=1awKeUpVP-GPlk6QmrAlmkSGzrNEW3uUN" #leer de internet
df_Data = pd.read_csv(url1, sep=',') #Leer archivo de url y separar por comas

#Importar el dataset de predicción a hacer
url = "https://drive.google.com/uc?export=download&id=1zLhk4TR1gkwZ5tLV3nnlxP21Pppl9H9p" #leer de internet
df_Final = pd.read_csv(url, sep=',') #Leer archivo de url y separar por coma

df_Data
```

```python
#reduccion manual del dataset
A = ['Id', 'BsmtFullBath', 'BsmtHalfBath', 'Alley', 'PoolQC', 'Fence', 'MiscFeature','FireplaceQu', 'HalfBath', 
               'LowQualFinSF', '2ndFlrSF', 'BsmtFinSF2', 'MasVnrArea', 'GarageType', 'GarageYrBlt', 'GarageFinish',
               'BsmtFinType2', 'BsmtQual', 'BsmtExposure', 'BsmtFinType1', 'MasVnrType']
df_Data.drop(A , axis='columns', inplace=True)

#Eliminar valores nulos
df_Data['LotFrontage'].fillna(0 , inplace=True)   #280 datos nullos reemplazados por 0
df_Data['GarageQual'].fillna('NA' , inplace=True)   #Reemplazar, Garajes no existentes
df_Data['GarageCond'].fillna('NA' , inplace=True)   #Reemplazar, Garajes no existentes
df_Data['Electrical'].fillna('Mix' , inplace=True)   #Reemplazar, Electricidad no existentes
df_Data['BsmtCond'].fillna('NA' , inplace=True)   #Reemplazar, Garajes no existentes


for i in df_Data.index:    
    if df_Data.loc[i,'LotArea'] >= 20000:      #Remover 5 datos atipicos
        df_Data.drop(index = i, axis = 'index', inplace=True)


#df_Data.describe()

#Verificación valores nulos

F = df_Data.isnull().sum()
print(F.to_string())
```

```python
#renombramiento de los features (columnas)
f = {"SepalLengthCm":"Sepalo_L", "Species":"Target",  "SepalWidthCm":"Sepalo_A", "PetalLengthCm":"Petalo_L", "PetalWidthCm":"Petalo_A"}

df_Data.rename(columns = f, inplace=True)             #renombramiento

df_Data
```

```python
#Splitting ("Division de los datos")
# train_test_split (dataset, % del entrenamiento)
dataset_train, dataset_test = train_test_split(df_Data, train_size=0.7, shuffle=True)

#verificar el data set
dataset_train

#verificar el test
dataset_test
```

```python
#separacion del Target y los features
#AHORA LOS features son X, el Target es Y
dataset_train_X = dataset_train.drop("Target", axis="columns")
dataset_train_Y = dataset_train["Target"]
dataset_test_X = dataset_test.drop("Target", axis="columns")
dataset_test_Y = dataset_test["Target"]

dataset_test_Y
```

```python
#Algoritmo de clasificacion (Supervisado)

#Naive Bayes Gausiano

#importar el modelo a ser utilizado
from sklearn.naive_bayes import GaussianNB  #Escoger el modelo de aprendizaje

#Instanciar el modelo
model = GaussianNB()

#model.fit Toma features y target
model.fit(dataset_train_X, dataset_train_Y)
```

```python
#Inferencia del modelo
inferencia = model.predict(dataset_test_X) #Prediccion del modelo
inferencia

#Validacion Automatica
from sklearn.metrics import accuracy_score
#la funcion accuracy_score toma como argumento (Target_Validation, Inferencia)
model_accuracy = accuracy_score(dataset_test_Y, inferencia)
print("Exactitud: ", str(model_accuracy*100),"%")
```

```python
#Decodificación de los datos

H = []

for i in range(len(infe)):
    if infe[i] == 0:
      H.append('Iris-versicolor')
    elif infe[i] == 1:
      H.append('Iris-setosa')
    elif infe[i] == 2:
      H.append('Iris-virginica')
    else:
      pass

J = []
for index in dataset_test_Y.index:
    if dataset_test_Y.loc[index] == 'Iris-versicolor':
        J.append('Iris-versicolor')
    elif dataset_test_Y.loc[index] == 'Iris-setosa':
        J.append('Iris-setosa')
    elif dataset_test_Y.loc[index] == 'Iris-virginica':
        J.append('Iris-virginica')
    else:
        pass

print("infe: ", infe)
print("--------------------------------------")
print("H :", H)
print("J :", J)
print("--------------------------------------")
print("Dataset: ", dataset_test_Y)
print("--------------------------------------")

#la funcion accuracy_score toma como argumento (Target_Validation, Inferencia)
model_accuracy = accuracy_score(J, H)
print("Exactitud: ", str(model_accuracy*100),"%")
```
